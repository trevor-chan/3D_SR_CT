```python
import pandas as pd 
import matplotlib.pyplot as plt
import os
from io import StringIO
import numpy as np
```

```python
file_path = os.path.join("/home/trevor/trevor/repos/3D_SR_CT/video_edm2/experiments/mnist_test", 'stats.jsonl')
```

```python
with open(file_path, 'r') as f:
    json_data = f.read()

df = pd.read_json(StringIO(json_data), lines=True)
```

```python
plt.figure(figsize = (10, 5))
plt.plot(df['Progress/kimg'], df['Loss/loss'])
plt.xlabel('Progress (kimg)')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.grid(True)
plt.show()
```
